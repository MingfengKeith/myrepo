```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
mpg = pd.read_csv('mpg.csv')

tag = ['4','f','r']
clr = {'f':'goldenrod','4':'lightblue','r':'black'}
label = {'f':'FWD','4':'4WD','r':'RWD'}
fig,ax = plt.subplots(nrows = 2, ncols = 2, figsize = (15,15))



for i in range(3):
    displ = mpg.loc[mpg['drv'] == tag[i]]['displ']
    print (len(displ))
    cty = mpg.loc[mpg['drv'] == tag[i]]['cty']
    displ3 = displ + np.random.rand(len(displ))*0.1
    cty3 = cty + np.random.rand(len(cty))*0.1 
    displ4 = displ + np.random.rand(len(displ))
    cty4 = cty + np.random.rand(len(cty))
    ax[0,0].scatter(displ,cty,label = label[tag[i]],marker = 'o',s=60,c=clr[tag[i]])
    ax[0,0].legend(loc = 0,bbox_to_anchor=[0.7, 0.7])
    ax[0,0].set_xlabel("displacement(l)")
    ax[0,0].set_ylabel(r"fuel economy(mpg)")
    ax[0,1].scatter(displ,cty,label = label[tag[i]],marker = 'o',s=60,c=clr[tag[i]],alpha = 0.5)
    ax[0,1].legend(loc = 0,bbox_to_anchor=[0.7, 0.7])
    ax[0,1].set_xlabel("displacement(l)")
    ax[0,1].set_ylabel(r"fuel economy(mpg)")
    ax[1,0].scatter(displ3,cty3,label = label[tag[i]],marker = 'o',s=60,c=clr[tag[i]],alpha = 0.5)
    ax[1,0].legend(loc = 0,bbox_to_anchor=[0.7, 0.7])
    ax[1,0].set_xlabel("displacement(l)")
    ax[1,0].set_ylabel(r"fuel economy(mpg)")
    ax[1,1].scatter(displ4,cty4,label = label[tag[i]],marker = 'o',s=60,c=clr[tag[i]],alpha = 0.5)
    ax[1,1].legend(loc = 0,bbox_to_anchor=[0.7, 0.7])
    ax[1,1].set_xlabel("displacement(l)")
    ax[1,1].set_ylabel(r"fuel economy(mpg)")
```
